```python
import pandas as pd
import numpy as np

def getData():

    energy = pd.read_excel('Energy Indicators.xls', header=None, skipfooter=2)
    #print (energy)
    energy = (energy.drop([0, 1], axis=1) 
                     .dropna() #drop NaN's
                     .drop(9)
                     .rename(columns={2: 'Country', 
                                      3: 'Energy Supply',
                                      4: 'Energy Supply per Capita',
                                      5: '% Renewable'})
                     .replace(regex=True, to_replace=[r'\d', r' \(([^)]+)\)'], value=r'') 
                     .replace(to_replace=["...", "Republic of Korea", "United States of America", 
                                         "United Kingdom of Great Britain and Northern Ireland",
                                         "China, Hong Kong Special Administrative Region"],
                              value=[np.NaN, "South Korea", "United States", "United Kingdom", "Hong Kong"])) 
    energy['Energy Supply'] = energy['Energy Supply'] * 1000000
    #energy.set_index(['Country'],inplace=True)
 
    GDP = pd.read_csv('world_bank.csv', header=None, skiprows=4)
    GDP = (GDP.rename(columns=GDP.iloc[0])
              .drop(0)
              .replace(to_replace=["Korea, Rep.", "Iran, Islamic Rep.", "Hong Kong SAR, China"],
                       value=["South Korea", "Iran", "Hong Kong"])
             .rename(columns={2006: '2006', 2007: '2007', 2008: '2008', 2009: '2009', 2010: '2010',
                                   2011: '2011', 2012: '2012', 2013: '2013', 2014: '2014', 2015: '2015'}))
   # GDP.set_index(['Country'],inplace=True) 
    ScimEn = pd.read_excel('scimagojr-3.xlsx')
    return energy, GDP, ScimEn

def answer_one():
    energy, GDP, ScimEn = getData()
    energy.to_excel
    energy = energy.dropna()
    GDP_columns = ['Country Name', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
    GDP = GDP[GDP_columns] 
    ScimEn_columns = ['Rank', 'Country', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index']
    ScimEn = ScimEn[ScimEn_columns]
    ScimEn = ScimEn[:15] 

    new = pd.merge(energy, GDP, how="inner", left_on="Country", right_on="Country Name")
    new = new.drop(['Country Name'], axis=1)
    new = pd.merge(new, ScimEn, how="inner", left_on='Country', right_on='Country')
    new = new.set_index('Country')
    columns = ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
    new = new[columns]
    return new
answer_one()

```

```python
def answer_two():
    energy, GDP, ScimEn = getData()
    m1 = pd.merge(energy, GDP, how="outer", left_on="Country", right_on="Country Name")
    m2 = pd.merge(m1, ScimEn, how="outer", left_on="Country", right_on="Country")
   # m3 = pd.merge(GDP, ScimEn, how="outer", left_on="Country Name", right_on="Country")
   # actual = pd.merge(m1, ScimEn, how="inner", left_on="Country", right_on="Country")
    g = len(m2.index)-len(answer_one())
    return g
answer_two()
```

```python
def answer_three():
    Top15 = answer_one()
    avgGDP = Top15.loc[:, '2006':'2015'].mean(axis=1).sort_values(ascending=False)
    return avgGDP
answer_three()
```

```python
def answer_four():
    Top15 = answer_one()
    country = answer_three().index[5]
    state2015 = Top15.loc[country, '2015']
    state2006 = Top15.loc[country, '2006']
    ans = state2015 - state2006
    return ans
answer_four()
```

```python
def answer_five():
    Top15 = answer_one()
    ans = Top15['Energy Supply per Capita'].mean()
    return ans
answer_five()
```

```python
def answer_six():
    Top15 = answer_one()
    state = Top15['% Renewable'].idxmax() 
    value = Top15['% Renewable'].max() 
    return state, value
answer_six()
```

```python
def answer_seven():
    Top15 = answer_one()
    Top15['Self-citations Ratio'] = Top15['Self-citations'] / Top15['Citations']
    state = Top15['Self-citations Ratio'].idxmax()
    value = Top15['Self-citations Ratio'].max()
    return state, value
answer_seven()
```

```python
def answer_eight():
    Top15 = answer_one()
    Top15['People'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    # Top15.dtypes
    Top15 = Top15.sort_values('People', ascending=False)
# Top15['Population']
# Top15
    state = Top15.index[2]
    return state
answer_eight()
```

```python
def answer_nine():
    Top15 = answer_one()
    Top15['People'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15['Citable documents per capita'] = Top15['Citable documents'] / Top15['People']
    ans = Top15['Citable documents per capita'].corr(Top15['Energy Supply per Capita'])
    return ans
answer_nine()
```

```python
def answer_ten():
    Top15 = answer_one()
    median_value = Top15['% Renewable'].median()
    states_renew = Top15['% Renewable'].map(lambda x: 1 if x >= median_value else 0)
    return states_renew
answer_ten()
```

```python

def answer_eleven():
    df = answer_one()
    df['people']=df['Energy Supply']/df['Energy Supply per Capita']
    df['people'] = np.float64(df['people'])
    ContinentDict = {'China':'Asia', 'United States':'North America', 'Japan':'Asia', 'United Kingdom':'Europe', 'Russian Federation':'Europe', 'Canada':'North America',
 'Germany':'Europe',
 'India':'Asia',
 'France':'Europe',
 'South Korea':'Asia',
 'Italy':'Europe',
 'Spain':'Europe',
 'Iran':'Asia',
 'Australia':'Australia',
 'Brazil':'South America'}
    df.rename(index=ContinentDict,inplace=True)
    df.reset_index(inplace = True)
    result = df[['Country','people']].groupby('Country').agg(['size', 'sum', 'mean', 'std'])
    return(result)
answer_eleven()

```

```python
def answer_twelve():
    ContinentDict = {'China':'Asia', 'United States':'North America', 'Japan':'Asia', 'United Kingdom':'Europe', 'Russian Federation':'Europe', 'Canada':'North America',
 'Germany':'Europe',
 'India':'Asia',
 'France':'Europe',
 'South Korea':'Asia',
 'Italy':'Europe',
 'Spain':'Europe',
 'Iran':'Asia',
 'Australia':'Australia',
 'Brazil':'South America'}
    Top15 = answer_one()
    Top15 = Top15.reset_index()
    Top15['% Renewable Bin'] = pd.cut(Top15['% Renewable'], bins=5, labels=["lowest", "low", "medium", "high", "highest"])
    Top15['Continent'] = Top15['Country'].map(ContinentDict)
    ans = Top15.groupby(['Continent', '% Renewable Bin'])['Country'].count()
    return ans
answer_twelve()
```

```python
def answer_thirteen():
    Top15 = answer_one()
    Top15['PopEst'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15['PopEst'] = Top15['PopEst'].map('{:,}'.format)
    return Top15['PopEst']
answer_thirteen()
```
